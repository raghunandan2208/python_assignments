```python
import datetime

class RentalShop:
    def __init__(self,stock = 0):
        self.stock = stock
        
    def display_stock(self):
        print("Bikes available on stock: ", self.stock)
        return self.stock
    
    def hourly_basis(self, num):
        
        if num <= 0:
            print("Enter Possitive integer")
            return None
        
        elif num > self.stock:
            print("currently available bikes: ", self.stock)
            return None
        
        else:
            time = datetime.datetime.now()
            print("Number of bikes taken for rent on hourly basis: ",num)
            print("Time of renting bike: ", time.hour)
            
            self.stock -= num
            return time
        
    def daily_basis(self, num):
            
        if num <= 0:
            print("Enter Possitive integer")
            return None
        
        elif num > self.stock:
            print("currently available bikes: ", self.stock)
            return None
        
        else:
            time = datetime.datetime.now()
            print("Number of bikes taken for rent on daily basis : ",num)
            print("Time of renting bike: ", time.hour)
            
            self.stock -= num
            return time 
        
    def weekly_basis(self, num):
            
        if num <= 0:
            print("Enter Possitive integer")
            return None
        
        elif num > self.stock:
            print("currently available bikes: ", self.stock)
            return None
        
        else:
            time = datetime.datetime.now()
            print("Number of bikes taken for rent on weekly basis : ",num)
            print("Time of renting bike: ", time.hour)
            
            self.stock -= num
            return time
        
    def return_bike(self, request):
        bikes, rent_basis, rent_time = request
        bill = 0
        if rent_time and rent_basis and bikes:
            self.stock += bikes
            time = datetime.datetime.now()
            total_time = time-rent_time
            
            if rent_basis ==1:
                bill = (total_time.seconds/3600)*5*bikes
            
            elif rent_basis == 2:
                bill = (total_time.days*20*bikes)
                
            elif rent_basis == 3:
                bill  = (total_time.days/7)*50*bikes
                
            if (bikes >= 3 and bikes <= 5):
                bill = bill * 0.7
                
            print("Your bill amount is : $",bill)
            
        return bill
    
class Customer:
    def __init__(self):
        self.bikes = 0
        self.rent_basis = 0
        self.rent_time = 0
        self.bill = 0
    
    def request_bike(self):
        bikes = int(input("Please enter the no of bikes u want to take rent: "))
        self.bikes = bikes
        return self.bikes
    
    def return_bike(self):
        if self.bikes and self.rent_basis and self.rent_time:
            return self.bikes, self.rent_basis, self.rent_time
        else:
            return 0,0,0
        
        
    

shop = RentalShop(50)
customer = Customer()


while True:
    print("Welcome to ZAKE Bike rental services")
    print("""
        1. Rent a bike on hourly basis $5
        2. Rent a bike on daily basis $20
        3. Rent a bike on weekly basis $60
        4. Return a bike
        5. Exit
        """)

    option = int(input("Please enter your option: "))

    shop.display_stock()


    if option == 1:
        customer.rent_time = shop.hourly_basis(customer.request_bike())
        customer.rent_basis = 1

    elif option == 2:
        customer.rent_time = shop.daily_basis(customer.request_bike())
        customer.rent_basis = 2

    elif option == 3:
        customer.rent_time = shop.weekly_basis(customer.request_bike())
        customer.rent_basis = 3

    elif option == 4:
        customer.bill = shop.return_bike(customer.return_bike())
        customer.bikes, customer.rent_basis, customer.rent_time = 0,0,0

    elif option == 5:
        print("Thank You for choosing our bike rental services")
        break

    else:
        print("Invalid option, please choose valid option...")
```
